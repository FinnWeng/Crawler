```python
def list_crawler(pg=5):
    url = "https://www.amazon.com/s/?rh=n%3A283155%2Cn%3A%211000%2Cn%3A17&page="
    headers = {
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36"
    }    
    
    article_data = []

    for p in range(1,pg+1):
        print("[INFO] crawling" + url + str(p))
        resp = requests.get(url+str(p),headers=headers)
        soup = BeautifulSoup(resp.text, 'html5lib')
        urls = [x.get('href') for x in soup.select('a.s-access-detail-page') if x.get('href').startswith('https')]
        for article_url in urls:
            urlshead = article_url.split("/ref")
            article_data.append(article_crawler(urlshead[0]))
                  
    return article_data 
```
